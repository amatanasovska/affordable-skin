```python
import pandas as pd
from pandasql import sqldf
```

```python
def filter_brand(words, list_to_filter):
    final_list = list(list_to_filter)
    for i in range(len(final_list)):
        for word in words:
            final_list[i] = "".join(final_list[i].split(word)).strip()

    return final_list
```

```python
def levenshtein_distance(s, t):
    # Get the length of the two input strings
    m, n = len(s), len(t)
    
    # Create a matrix of size (m+1)x(n+1) to store the distances
    distance = [[0 for j in range(n+1)] for i in range(m+1)]
    
    # Fill in the first row and column of the matrix
    for i in range(1, m+1):
        distance[i][0] = i
    for j in range(1, n+1):
        distance[0][j] = j
    
    # Compute the Levenshtein distance
    for j in range(1, n+1):
        for i in range(1, m+1):
            if s[i-1] == t[j-1]:
                cost = 0
            else:
                cost = 1
            distance[i][j] = min(distance[i-1][j] + 1, distance[i][j-1] + 1, distance[i-1][j-1] + cost)
    
    # Return the Levenshtein distance
    return distance[m][n]

```

### Find the same products using levenshtein distance

```python
equal_rows = []
```

```python
found_their_equal = []
```

```python
for index_y,row_y in df_yeppeuda_not_present.iterrows():
    for index_c,row_c in df_coslovemetics_not_present.iterrows():
        title_y = row_y['title']
        title_c = row_c['title']
        dist = levenshtein_distance(title_y.lower(),title_c.lower())
        
        if dist<max(len(title_y),len(title_c))//3:
            equal_rows.append((row_y,row_c))
            found_their_equal.append(row_y)
            found_their_equal.append(row_c)
```
