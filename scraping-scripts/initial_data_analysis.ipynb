```python
import pandas as pd
from pandasql import sqldf
```

```python
def filter_brand(words, list_to_filter):
    final_list = list(list_to_filter)
    for i in range(len(final_list)):
        for word in words:
            final_list[i] = "".join(final_list[i].split(word)).strip()

    return final_list
```

```python
def levenshtein_distance(s, t):
    # Get the length of the two input strings
    m, n = len(s), len(t)
    
    # Create a matrix of size (m+1)x(n+1) to store the distances
    distance = [[0 for j in range(n+1)] for i in range(m+1)]
    
    # Fill in the first row and column of the matrix
    for i in range(1, m+1):
        distance[i][0] = i
    for j in range(1, n+1):
        distance[0][j] = j
    
    # Compute the Levenshtein distance
    for j in range(1, n+1):
        for i in range(1, m+1):
            if s[i-1] == t[j-1]:
                cost = 0
            else:
                cost = 1
            distance[i][j] = min(distance[i-1][j] + 1, distance[i][j-1] + 1, distance[i-1][j-1] + cost)
    
    # Return the Levenshtein distance
    return distance[m][n]

```

```python
df_yeppeuda = pd.read_csv("yeppeuda_tiam.csv")
```

```python
df_coslovemetics = pd.read_csv("coslovemetics_tiam.csv")
```

```python
q = """select * from df_yeppeuda as dy join df_coslovemetics as dc on dy.title=dc.title"""
df = sqldf(q, globals())
```

```python
df_yeppeuda['title']= filter_brand(["TIA'M - ", "TIAM"],df_yeppeuda['title'])
```

```python
df_coslovemetics['title']= filter_brand(["TIA'M - ", "TIAM"],df_coslovemetics['title'])
```

```python
q = """select * from df_yeppeuda as dy join df_coslovemetics as dc on dy.title=dc.title"""
df_after_clear = sqldf(q, globals())
```

```python
df_after_clear
```

```python
q = """select * from df_yeppeuda as dy where not exists
    (
        select * from df_after_clear as dac where dac.title=dy.title
    )"""
df_yeppeuda_not_present = sqldf(q, globals())
```

```python
df_yeppeuda_not_present
```

```python
q = """select * from df_coslovemetics as dc where not exists
    (
        select * from df_after_clear as dac where dac.title=dc.title
    )"""
df_coslovemetics_not_present = sqldf(q, globals())
```

```python
df_coslovemetics_not_present
```

```python
equal_rows = []
```

```python
for index_y,row_y in df_yeppeuda_not_present.iterrows():
    for index_c,row_c in df_coslovemetics_not_present.iterrows():
        title_y = row_y['title']
        title_c = row_c['title']
        dist = levenshtein_distance(title_y,title_c)

        if dist<len(title_y)//2:
            equal_rows.append((row_y,row_c))
```
